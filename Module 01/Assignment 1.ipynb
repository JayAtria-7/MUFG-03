```python
import numpy as np

def calculate_iqr(data):
    """
    Calculates the Interquartile Range (IQR) for a given dataset.

    The IQR is the difference between the 75th percentile (Q3) and the
    25th percentile (Q1). It's a measure of statistical dispersion.

    Args:
        data: A list or numpy array of numerical data.

    Returns:
        The IQR value as a float, or None if the data is empty.
    """
    if not data:
        print("Error: Input data cannot be empty.")
        return None

    # Calculate the first quartile (25th percentile)
    q1 = np.percentile(data, 25)

    # Calculate the third quartile (75th percentile)
    q3 = np.percentile(data, 75)

    # IQR is the difference between Q3 and Q1
    iqr = q3 - q1
    return iqr

def calculate_z_scores(data):
    """
    Calculates the Z-score for each data point in a dataset.

    A Z-score indicates how many standard deviations an element is
    from the mean.

    Args:
        data: A list or numpy array of numerical data.

    Returns:
        A list of Z-scores corresponding to each data point,
        or an empty list if the data is empty or has no deviation.
    """
    if not data:
        print("Error: Input data cannot be empty.")
        return []

    mean = np.mean(data)
    std_dev = np.std(data)

    # Avoid division by zero if all data points are the same
    if std_dev == 0:
        print("Warning: Standard deviation is zero. Z-scores will all be 0.")
        return [0] * len(data)

    # Calculate Z-score for each data point
    z_scores = [(x - mean) / std_dev for x in data]
    return z_scores

# --- Example Usage ---
if __name__ == "__main__":
    # Sample dataset
    sample_data = [2, 3, 5, 6, 7, 8, 8, 9, 10, 12, 15, 20, 50]

    print(f"Sample Data: {sample_data}\n")

    # --- Calculate and print IQR ---
    print("--- Interquartile Range (IQR) ---")
    iqr_value = calculate_iqr(sample_data)
    if iqr_value is not None:
        q1 = np.percentile(sample_data, 25)
        q3 = np.percentile(sample_data, 75)
        print(f"Q1 (25th percentile): {q1}")
        print(f"Q3 (75th percentile): {q3}")
        print(f"IQR (Q3 - Q1): {iqr_value}")
        # Identify outliers using the 1.5*IQR rule
        lower_bound = q1 - 1.5 * iqr_value
        upper_bound = q3 + 1.5 * iqr_value
        outliers = [x for x in sample_data if x < lower_bound or x > upper_bound]
        print(f"Potential outliers (outside {lower_bound:.2f} and {upper_bound:.2f}): {outliers}\n")


    # --- Calculate and print Z-scores ---
    print("--- Z-Scores ---")
    z_score_values = calculate_z_scores(sample_data)
    if z_score_values:
        print("Data Point | Z-Score")
        print("-----------------------")
        for i in range(len(sample_data)):
            print(f"{sample_data[i]:<10} | {z_score_values[i]:.4f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def calculate_iqr(data):
    """
    Calculates the Interquartile Range (IQR) for a given dataset.

    The IQR is the difference between the 75th percentile (Q3) and the
    25th percentile (Q1). It's a measure of statistical dispersion.

    Args:
        data: A list or numpy array of numerical data.

    Returns:
        The IQR value as a float, or None if the data is empty.
    """
    if data.empty if isinstance(data, pd.Series) else not any(data):
        print("Error: Input data cannot be empty.")
        return None

    # Calculate the first quartile (25th percentile)
    q1 = np.percentile(data, 25)

    # Calculate the third quartile (75th percentile)
    q3 = np.percentile(data, 75)

    # IQR is the difference between Q3 and Q1
    iqr = q3 - q1
    return iqr

def calculate_z_scores(data):
    """
    Calculates the Z-score for each data point in a dataset.

    A Z-score indicates how many standard deviations an element is
    from the mean.

    Args:
        data: A list or numpy array of numerical data.

    Returns:
        A list of Z-scores corresponding to each data point,
        or an empty list if the data is empty or has no deviation.
    """
    if data.empty if isinstance(data, pd.Series) else not any(data):
        print("Error: Input data cannot be empty.")
        return []

    mean = np.mean(data)
    std_dev = np.std(data)

    # Avoid division by zero if all data points are the same
    if std_dev == 0:
        print("Warning: Standard deviation is zero. Z-scores will all be 0.")
        return pd.Series([0] * len(data), index=data.index) if isinstance(data, pd.Series) else [0] * len(data)

    # Calculate Z-score for each data point
    z_scores = (data - mean) / std_dev
    return z_scores

def load_and_preprocess_data(filepath):
    """
    Loads data from a CSV file and performs basic preprocessing.

    Args:
        filepath: The path to the CSV file.

    Returns:
        A pandas DataFrame with preprocessed data, or None if file not found.
    """
    if not os.path.exists(filepath):
        print(f"Error: File not found at '{filepath}'")
        return None

    print(f"Loading data from '{filepath}'...")
    df = pd.read_csv(filepath)
    print("Original Data Head:")
    print(df.head())
    print("\nChecking for missing values...")
    print(df.isnull().sum())

    # --- Preprocessing Step: Handle missing values ---
    # For each numeric column with missing values, fill them with the mean
    for col in df.select_dtypes(include=np.number).columns:
        if df[col].isnull().any():
            print(f"Filling missing values in column '{col}' with the mean.")
            mean_val = df[col].mean()
            df[col].fillna(mean_val, inplace=True)

    print("\nPreprocessed Data Head:")
    print(df.head())
    return df

def perform_eda(df, column_name):
    """
    Performs Exploratory Data Analysis (EDA) on a specific column.

    Args:
        df: The pandas DataFrame.
        column_name: The name of the column to analyze.
    """
    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in the DataFrame.")
        return

    data = df[column_name]

    print(f"\n--- Exploratory Data Analysis (EDA) for '{column_name}' ---")

    # 1. Descriptive Statistics
    print("\nDescriptive Statistics:")
    print(data.describe())

    # 2. Visualization
    print("\nGenerating visualizations...")
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Histogram to show distribution
    sns.histplot(data, kde=True, ax=axes[0])
    axes[0].set_title(f'Distribution of {column_name}')
    axes[0].set_xlabel(column_name)
    axes[0].set_ylabel('Frequency')

    # Box plot to identify outliers
    sns.boxplot(y=data, ax=axes[1])
    axes[1].set_title(f'Box Plot of {column_name}')
    axes[1].set_ylabel(column_name)

    plt.suptitle(f'EDA Visualizations for {column_name}', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()


# --- Example Usage ---
if __name__ == "__main__":
    # 1. Create a dummy dataset for demonstration
    data_dict = {
        'ID': range(1, 21),
        'Score': [85, 92, 78, 88, 95, 65, 72, np.nan, 89, 91, 76, 83, 99, 68, 75, 81, 150, 93, 70, 80]
    }
    dummy_df = pd.DataFrame(data_dict)
    file_path = "sample_dataset.csv"
    dummy_df.to_csv(file_path, index=False)
    print(f"Created a dummy dataset '{file_path}' for demonstration.\n")

    # 2. Load and preprocess the data
    preprocessed_df = load_and_preprocess_data(file_path)

    if preprocessed_df is not None:
        analysis_column = 'Score'

        # 3. Perform EDA
        perform_eda(preprocessed_df, analysis_column)
        
        # Get the specific column data for further analysis
        scores = preprocessed_df[analysis_column]

        # 4. Calculate and print IQR
        print(f"\n--- Interquartile Range (IQR) for '{analysis_column}' ---")
        iqr_value = calculate_iqr(scores)
        if iqr_value is not None:
            q1 = np.percentile(scores, 25)
            q3 = np.percentile(scores, 75)
            print(f"Q1 (25th percentile): {q1:.2f}")
            print(f"Q3 (75th percentile): {q3:.2f}")
            print(f"IQR (Q3 - Q1): {iqr_value:.2f}")
            # Identify outliers using the 1.5*IQR rule
            lower_bound = q1 - 1.5 * iqr_value
            upper_bound = q3 + 1.5 * iqr_value
            outliers = scores[(scores < lower_bound) | (scores > upper_bound)]
            print(f"Potential outliers (outside {lower_bound:.2f} and {upper_bound:.2f}): {list(outliers)}\n")

        # 5. Calculate and print Z-scores
        print(f"--- Z-Scores for '{analysis_column}' ---")
        z_score_values = calculate_z_scores(scores)
        if not z_score_values.empty:
            z_scores_df = pd.DataFrame({'Score': scores, 'Z-Score': z_score_values})
            print(z_scores_df)

    # Clean up the dummy file
    os.remove(file_path)
    print(f"\nCleaned up dummy file: '{file_path}'")

```
